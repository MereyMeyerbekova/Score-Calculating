```python
import pandas as pd
roster = pd.read_csv('roster.csv')
roster
```

```python
roster[['Last Name', 'First Name']] = roster.Name.str.split(",",expand=True,)
del roster["Name"]
```

```python
roster.drop(['NetID', 'Section'], axis=1, inplace=True)
```

```python
roster.drop(['First Name', 'Last Name'], axis=1, inplace=True)
roster
```

```python
roster['Email Address'] = roster['Email Address'].str.lower()
dict = {'Email Address':'Email'}
roster.rename(columns=dict,inplace=True)
display(roster)
```

```python
df1 = pd.read_csv('quiz_1_grades.csv')
```

```python
df2 = pd.read_csv('quiz_2_grades.csv')
```

```python
df3 = pd.read_csv('quiz_3_grades.csv')
```

```python
df4 = pd.read_csv('quiz_4_grades.csv')
```

```python
df5 = pd.read_csv('quiz_5_grades.csv')
```

```python
merge1 = pd.merge(pd.merge(df1,df2,on=['Last Name', 'First Name', 'Email']),df3,on= ['Last Name', 'First Name', 'Email']) 
merge1
```

```python
dict = {'Grade_x':'Quiz1',
        'Grade_y':'Quiz2',
        'Grade':'Quiz3'}
 
merge1.rename(columns=dict,inplace=True) 
```

```python
merge2 = pd.merge(pd.merge(merge1, df4, on=['Last Name', 'First Name', 'Email']),df5,on= ['Last Name', 'First Name', 'Email'])
```

```python
dict = {'Grade_x':'Quiz4',
        'Grade_y':'Quiz5'}
merge2.rename(columns=dict,inplace=True)
display(merge2)
```

```python
hw_exam = pd.read_csv('hw_exam_grades.csv')
```

```python
import numpy as np
```

```python
merge3 = pd.merge(merge2, hw_exam, on=['Last Name', 'First Name'], how= 'left').fillna(0) 
merge3.replace(np.nan,0)
merge3
```

```python
merge3 = merge3.loc[:, ~merge3.columns.str.contains("Submission Time")]
del merge3["SID"]
merge3
```

```python
pd.options.mode.chained_assignment = None
```

```python
exam = 3
for n in range(1, exam + 1):
    merge3[f"Exam {n} Score"] = (merge3[f"Exam {n}"] / merge3[f"Exam {n} - Max Points"])
```

```python
hw_scores = merge3.filter(regex=r"^Homework \d\d?$", axis=1)
hw_max = merge3.filter(regex=r"^Homework \d\d? -", axis=1)
sum_hw_scores = hw_scores.sum(axis=1)
sum_max = hw_max.sum(axis=1)
merge3["Homework score"] = sum_hw_scores / sum_max
merge3
```

```python
maxq1 = merge3['Quiz1'].max()
maxq2 = merge3['Quiz2'].max()
maxq3 = merge3['Quiz3'].max()
maxq4 = merge3['Quiz4'].max()
maxq5 = merge3['Quiz5'].max()
print(maxq1, maxq2,maxq3, maxq4, maxq5)
```

```python
quiz = merge3.filter(regex=r"^Quiz[0-9]+$", axis=1)
quiz_max = [11, 15, 17, 14, 12]
sum_quiz = quiz.sum(axis=1)
sum_max = sum(quiz_max)
merge3["Quizes score"] = sum_quiz / sum_max
merge3
```

```python
grading = pd.Series({"Exam 1 Score": 0.10, "Exam 2 Score": 0.10, "Exam 3 Score": 0.10, "Quizes score": 0.30, "Homework score": 0.4, })
```

```python
merge3["Total Score"] = np.ceil(((merge3[grading.index] * grading).sum(axis=1))*100)
merge3
```

```python
merge3.to_csv('final_grades.csv')
```
